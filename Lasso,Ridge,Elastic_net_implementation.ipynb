```python
# since we have not done outlier treatment ,then the better idea would be replace missing value with median

df['horsepower'].median()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
# data encoding >> convert catgoricval into numerical

df['origin'] = df['origin'].map({'usa':1,'japan':2,'europe':3})
```

```python
df
```

```python
df.info()
```

```python
# separate into x and y

x = df.drop('mpg',axis = 1)
y = df['mpg']
```

```python
x
```

```python
y
```

```python
# train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

```python
x_train
```

```python
x_test
```

```python
# simple linear regression model

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# another method to coefficient

for i, col_name in enumerate (x_train.columns):
  print(f'The coefficient for {col_name} is {model.coef_[i]}')
```

```python
# these features might not be contributing in model training

from sklearn.metrics import r2_score
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
# Regularised model
# Ridge Regression

from sklearn.linear_model import Ridge
```

```python
ridge_regression_model = Ridge(alpha = 0.1)
```

```python
ridge_regression_model
```

```python
ridge_regression_model.fit(x_train,y_train)
```

```python
for i, col_name in enumerate (x_train.columns):
  print(f'The coefficient for {col_name} is {ridge_regression_model.coef_[i]}')
```

```python
# ridge reg evaluation

y_pred_ridge = ridge_regression_model.predict(x_test)
r2_score_ridge = r2_score(y_test,y_pred_ridge)
print(f'The r2 score of ridge regression model is {r2_score_ridge}')
```

```python
# Lasso Regression

from sklearn.linear_model import Lasso
```

```python
lasso_regression_model =  Lasso(alpha = 0.1)
```

```python
lasso_regression_model
```

```python
 lasso_regression_model.fit(x_train,y_train)
```

```python
for i, col_name in enumerate (x_train.columns):
  print(f'The coefficient for {col_name} is {lasso_regression_model.coef_[i]}')
```

```python
y_pred_lasso = lasso_regression_model.predict(x_test)
r2_score_lasso = r2_score(y_test,y_pred_lasso)
print(f'The r2 score of lasso regression model is {r2_score_lasso}')
```

```python
# Elastic net

from sklearn.linear_model import ElasticNet
```

```python
elastic_net_model = ElasticNet(alpha = 0.1,l1_ratio=.5)
```

```python
elastic_net_model.fit(x_train,y_train)
```

```python
for i, col_name in enumerate (x_train.columns):
  print(f'The coefficient for {col_name} is {elastic_net_model.coef_[i]}')
```

```python
y_pred_lasso =elastic_net_model.predict(x_test)
r2_score_elastic = r2_score(y_test,y_pred_lasso)
print(f'The r2 score of elastic net model is {r2_score_elastic}')
```
